### Deteksi Garis Tepi

```python
import cv2  # Mengimpor library OpenCV untuk pengolahan gambar dan video
import numpy as np  # Mengimpor library NumPy untuk manipulasi array
import matplotlib.pyplot as plt  # Mengimpor library Matplotlib untuk visualisasi
%matplotlib inline  # Memastikan plot Matplotlib ditampilkan di notebook Jupyter secara inline
import skimage  # Mengimpor library scikit-image untuk pemrosesan citra
```

```python
image = cv2.imread('1.jpg')  # Membaca gambar '1.jpg' dan menyimpannya ke dalam variabel image menggunakan OpenCV

cv2.imshow("Gambar Parkir", image)  # Menampilkan gambar dalam jendela dengan judul "Gambar Parkir"
cv2.waitKey(0)  # Menunggu tombol keyboard ditekan tanpa batasan waktu (0)
cv2.destroyAllWindows()  # Menutup semua jendela OpenCV setelah pengguna menekan tombol keyboard
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Mengubah gambar menjadi citra grayscale menggunakan OpenCV
edges = cv2.Canny(image, 100, 150)  # Mendeteksi tepi gambar menggunakan metode Canny dengan threshold 100 dan 150
```

```python
cv2.imshow("Gambar Parkir", edges)  # Menampilkan citra hasil deteksi tepi dalam jendela dengan judul "Gambar Parkir"
cv2.waitKey(0)  # Menunggu tombol keyboard ditekan tanpa batasan waktu (0)
cv2.destroyAllWindows()  # Menutup semua jendela OpenCV setelah pengguna menekan tombol keyboard
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 10))  # Membuat subplot dengan 1 baris dan 2 kolom, dengan ukuran gambar 10x10 inci
ax = axs.ravel()  # Meratakan array dari objek sumbu (axes)

ax[0].imshow(gray, cmap="gray")  # Menampilkan gambar grayscale dalam subplot pertama
ax[0].set_title("Gambar Asli")  # Memberi judul subplot pertama

ax[1].imshow(edges, cmap="gray")  # Menampilkan gambar hasil deteksi tepi dalam subplot kedua
ax[1].set_title("Gambar Setelah diolah")  # Memberi judul subplot kedua

plt.show()  # Menampilkan plot Matplotlib yang sudah dibuat
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap=20)  # Mendeteksi garis-garis menggunakan transformasi Hough probabilistik pada citra tepi
image_line = image.copy()  # Membuat salinan gambar asli untuk digunakan dalam menampilkan garis-garis deteksi
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]  # Mendapatkan koordinat titik awal dan akhir garis dari setiap garis yang terdeteksi
    cv2.line(image_line, (x1, y1), (x2, y2), (100, 8, 255), 1)  # Menggambar garis pada gambar salinan dengan warna (100, 8, 255) dan ketebalan garis 1
```

```python
fig,axs = plt.subplots(1,2,figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(gray, cmap = "gray")
ax[0].set_title("Gambar Asli")

ax[1].imshow(image_line, cmap = "gray")
ax[1].set_title("Gambar setelah diolah")
```
